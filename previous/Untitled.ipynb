```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
size = 5
x = np.arange(1, size+1)
a = np.arange(size)
b = np.arange(size)
c = np.arange(size)

total_width, n = 0.8, 3
width = total_width / n
x = x - (total_width - width) / 2

plt.bar(x, a,  width=width, label='a')
plt.bar(x + width, b, width=width, label='b')
plt.bar(x + 2 * width, c, width=width, label='c')

plt.tick_params(axis='x',length=0)
plt.legend()
plt.show()
```

```python
x = np.arange(0,len(df['date'])*2,2)
width=0.5
x1 = x-width/2
x2 = x+width/2
y1 = df['beijing']
y2 = df['guangzhou']

# 绘制分组柱状图

plt.bar(x1,y1,width=0.5,label='beijing',color='#f9766e',edgecolor='k',zorder=2)
plt.bar(x2,y2,width=0.5,label='guangzhou',color='#00bfc4',edgecolor='k',zorder=2)

# 添加x,y轴名称、图例和网格线
plt.xlabel('2022',fontsize=11)
plt.ylabel('AQI',fontsize=11)
plt.legend(frameon=False)
# plt.grid(ls='--',alpha=0.8)

# 修改x刻度标签为对应日期
plt.xticks(x,df['date'],fontsize=10)
plt.tick_params(axis='x',length=0)

plt.tight_layout()
plt.savefig('./bar2.png',dpi=600)
plt.show()
```
